# 5.  Feature Selction and Extration


*   Now in our data frame there are 262 features.
*   Now we need to select the most correlated features out of them.
*   To select these features, we need to calculate the correlation value between each feature and the label ('salary_in_usd').\
$$ Correlation Coefficient = \frac{\sum((x - \bar{x})(y - \bar{y}))}{\sqrt{\sum(x - \bar{x})^2} \cdot \sqrt{\sum(y - \bar{y})^2}} $$

$x$ - values of the x-variable in a sample \
$y$ - values of the y-variable in a sample \
$\bar{x}, \bar{y} $ = Mean

```python
target_correlation = data_clean.corr()['salary_in_usd'].apply(abs).sort_values()
target_correlation
```

```python
feature_selection = {}
i = 0
for index, corrleation in target_correlation.items():
  if corrleation > 0.1:
    i += 1 
    feature_selection[index]=corrleation
i
print(i)
feature_selection = dict(sorted(feature_selection.items(), key=lambda x: x[1]))
feature_selection
```

According to this correlation values of each features in our Data Frame, maximum correlation value is0.5143300491244743 It suggests that there is a weak **linear relationship** between our input features and the target variable.

```python
limited_df = data_clean[list(sorted(feature_selection.keys()))]
limited_df
```

## Unsupervised Learning

### K - Means


*   K - Means is Partitioning clustring method.
*   K - means alogrithm is carried out three steps.
*   Follawing flow chart explain how K means work with 3 steps



image.png

To find best number of clusters. We can use Elbow method. We can plot a graph 'WCSS' vs 'K values' to find the best K. \
( WCSS - Within the Cluster sum of Squares )

```python
#@title Model
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
k_values = range(1, 11)  # Range of k values to evaluate
wcss_values = []  # List to store the WCSS values

for k in k_values:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(data_clean)
    wcss = kmeans.inertia_
    wcss_values.append(wcss)

plt.plot(k_values, wcss_values)
plt.title("WCSS vs. Number of Clusters")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("WCSS")
plt.show()
```

To measure the quality of K - Means alogorithem. We can use Silhouette Coefficient

```python
#@title Evaluation
kmeans = KMeans(n_clusters=4)
kmeans.fit(data_clean)
labels = kmeans.labels_
centers = kmeans.cluster_centers_
silhouette_avg = silhouette_score(limited_df, labels)
print('Silhouette Average :',silhouette_avg)
print('Centers of Clusters :', centers)
```

There are various ways to to select best Hyperparameter for a algorithm. Following methods are some of them.

*   Grid Search -  Trained and evaluated for all possible combinations of these  predefined set of values of each hyperparameter.

  *  Pros - Can find the best set of hyperparameters
  *  Cons - Takes more time


*   Random Search -  Random search selects random combinations of hyperparameter values from a predefined set of values.
  *  Pros -  Efficient than  Grid Search method
  *  Cons -Low chance to find the best set of hyperparameters than Grid Search method

```python
#@title Optimization
learning_rates = [0.1, 0.01, 0.001]

optimizers_list = [
    optimizers.SGD, 
    optimizers.Adam,
    optimizers.RMSprop
]

best_mse = float('inf')
best_model = None
best_lr = None
best_optimizer = None

for lr in learning_rates:
    for optimizer in optimizers_list:
        # Create the model
        model = keras.Sequential([
            layers.Dense(64, activation='relu', input_shape=(X_train.shape[1],)),
            layers.Dense(64, activation='relu'),
            layers.Dense(1)
        ])

        model.compile(optimizer=optimizer(learning_rate=lr), loss='mean_squared_error')

        model.fit(X_train, y_train, batch_size=32, epochs=100, verbose=0)

        mse = model.evaluate(X_test, y_test, verbose=0)

        if mse < best_mse:
            best_mse = mse
            best_model = model
            best_lr = lr
            best_optimizer = optimizer.__name__

print("Best Combination - Learning Rate:", best_lr)
print("Best Combination - Optimizer:", best_optimizer)
print("Best Combination - Mean Squared Error:", best_mse)
```
